```python
def detect_text_local(file_path_string, MIN_CONF=30, MIN_LEN=3):
    from io import BytesIO
    from PIL import Image
    import boto3
    session = boto3.Session(profile_name='default')
    client = session.client('rekognition')

    img = Image.open(file_path_string)
    byte_io = BytesIO()
    img.save(byte_io, 'JPEG')
    img_binary = byte_io.getvalue()

    response = client.detect_text(Image={'Bytes':img_binary})
    textDetections = response['TextDetections']

    words = [ (i['DetectedText'].lower(),i['Confidence'])  for i in textDetections if i['Type'] == 'WORD' 
                                                    and i['Confidence'] > MIN_CONF 
                                                    and len(i['DetectedText']) >= MIN_LEN 
            ]
    
    img.save(f'./images_w_labels/{words[0][0]}.jpg')
    #return words
```

```python
for i in range(2):
    detect_text_local(f'./images/image_{i}.jpg')
```

```python
import threading

threads = []
for i in range(3000): 
    thread = threading.Thread(target=detect_text_local, args=(f'./images/image_{i}.jpg',))
    thread.start()
    threads.append(thread)

for thread in threads:
    thread.join()
```
